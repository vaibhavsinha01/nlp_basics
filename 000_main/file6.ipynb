```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

cv = CountVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm = cv.fit_transform(npr['Article'])
print(dtm)

LDA = LatentDirichletAllocation(n_components=7,random_state=42)
LDA.fit(dtm)
```

```python
import random
print(len(cv.get_feature_names_out()))
print(type(cv.get_feature_names_out()))
print(cv.get_feature_names_out()[41000])

random_word_id = random.randint(0,54777)
cv.get_feature_names_out()[random_word_id]

print(len(LDA.components_))
print(type(LDA.components_))
print('\n')
print(LDA.components_.shape)
LDA.components_
```

```python
single_topic = LDA.components_[0]
print(single_topic.argsort())

# arr = np.array([10,200,1])
# print(arr.argsort())

single_topic.argsort()[:-10]

top_twenty_words = single_topic.argsort()[-20:]
print(top_twenty_words)

for index in top_twenty_words:
    print(cv.get_feature_names_out()[index])
```

```python
for i,topic in enumerate(LDA.components_):
    print(f'The top 15 words for the topic #{i}')
    print([cv.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
```

```python
topic_results = LDA.transform(dtm)
print(topic_results[0])
print(topic_results[0].round(2))
print(topic_results[0].argmax())
topic_results.shape

npr['Topic'] = topic_results.argmax(axis=1)
print(npr)
```

```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

npr = pd.read_csv(r'C:\Users\Vaibhav\OneDrive\Documents\FolderPython\Artificial_Intelligence\Natural_Language_Processing\UPDATED_NLP_COURSE\05-Topic-Modeling\npr.csv')
tfidf = TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')

dtm = tfidf.fit_transform(npr['Article'])
print(dtm)
```

```python
nmf_model = NMF(n_components=7,random_state=42)
nmf_model.fit(dtm)
tfidf.get_feature_names_out()[2300]

for index,topic in enumerate(nmf_model.components_):
    print(f'The top 15 words for the topic are #{index}')
    print([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-15:]])
    print('\n')
```

```python
topic_results = nmf_model.transform(dtm)
topic_results[0].argmax

npr['Topic'] = topic_results.argmax(axis=1)
my_topic_dict = {
    0: 'Politics',
    1: 'Economy',
    2: 'Sports',
    3: 'Health',
    4: 'Technology',
    5: 'Education',
    6: 'Environment'
}
npr.head()
npr['Topic Label']=npr['Topic'].map(my_topic_dict)
print(npr)
```
