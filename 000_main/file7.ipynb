```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
type(iris)
print(iris.DESCR)
```

```python
x = iris.data
y = iris.target

from keras.utils import to_categorical

y = to_categorical(y)
print(y)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense,Dropout

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
scaler = StandardScaler()

scaler.fit_transform(x_train)
scaler.transform(x_test)

model = Sequential()
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(8,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(3,activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()

model.fit(x_train,y_train,epochs=1000,verbose=2)
```

```python
predictions = np.argmax(model.predict(x_test),axis=1)
print(predictions)

print(np.argmax(y_test,axis=1))

from sklearn.metrics import confusion_matrix,classification_report
from keras.models import load_model

# code for saving the model

# new_model = load_model('myfirstmodel.h5')

print(confusion_matrix(y_test.argmax(axis=1),predictions))
print(classification_report(y_test.argmax(axis=1),predictions))
```

```python
def read_file(filepath):
    with open(filepath) as f:
        str_text = f.read()
    return str_text
```

```python
import spacy

nlp = spacy.load('en_core_web_lg',disable=['parser','tagger','ner'])
nlp.max_length = 1198623

def seperate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']

d = read_file(r'C:\Users\Vaibhav\OneDrive\Documents\FolderPython\Artificial_Intelligence\Natural_Language_Processing\Nlp_revision\06-Deep-Learning\moby_dick_four_chapters.txt')
tokens = seperate_punc(d)
tokens
```

```python
train_len = 25+1

text_sequences = []

for i in range(train_len,len(tokens)):
    seq = tokens[i-train_len:i]
    text_sequences.append(seq)

type(text_sequences)
print(text_sequences[0])

print(' '.join(text_sequences[0]))
print(' '.join(text_sequences[1]))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

# initialize the tokenizer
tokenizer = Tokenizer()

# assuming text sequence is your list of text sequences
tokenizer.fit_on_texts(text_sequences)

# convert text sequences to numerical sequences
sequences = tokenizer.texts_to_sequences(text_sequences)

print(sequences)
print(tokenizer.index_word)

for i in sequences[0]:
    print(f'{i}:{tokenizer.index_word[i]}:word count {tokenizer.word_counts}')
```

```python
vocabulary_size = len(tokenizer.word_counts)
print(vocabulary_size)
print(type(sequences))

import numpy as np
sequences = np.array(sequences)
print(sequences)
```

```python
from keras.utils import to_categorical

x = sequences[:,:-1]
y = sequences[:,-1]
y = to_categorical(y,num_classes=vocabulary_size+1)

seq_len = x.shape[1]
seq_len
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM,Embedding

def create_model(vocabulary_size,seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size,25,input_length=seq_len))
    model.add(LSTM(150,return_sequences=True))
    model.add(LSTM(150))
    model.add(Dense(150,activation='relu'))
    model.add(Dense(vocabulary_size,activation='softmax'))
    model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
    model.summary()
    return model

model = create_model(vocabulary_size+1,seq_len)
```

```python
from pickle import dump,load

model.fit(x,y,batch_size=128,epochs=20,verbose=1)
model.save('my_mobydick_model.h5')
dump(tokenizer,open('my_simple_tokenizer','wb'))
```

```python
from keras.preprocessing.sequence import pad_sequences

def generate_text(model,tokenizer,seq_len,seed_text,num_gen_words):
    output_text = []
    input_text = seed_text

    for i in range(num_gen_words):
        encoded_text = tokenizer.texts_to_sequence([input_text])[0]
        pad_encoding = pad_sequences([encoded_text],maxlen=seq_len,truncating='pre')
        pred_probs = model.predict(pad_encoding,verbose=0)
        pred_word_ind = np.argmax(pred_probs,axis=-1)[0]
        pred_word = tokenizer.index_word.get(pred_word_ind,'')
        input_text += " "+pred_word
        output_text.append(pred_word)
        
    return ' '.join(output_text)
```
